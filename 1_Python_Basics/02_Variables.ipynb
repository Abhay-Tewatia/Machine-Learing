```python
# Understand variables and their types
# python is dynamically typed language, which means you don't need to declare the type of a variable when you create it. The type is inferred based on the value assigned to the variable.
# You can change the type of a variable by assigning a different type of value to it.
my_variable = 10 # my_variable is an integer
print("my_variable: ", my_variable, " Type: ", type(my_variable)) 
my_variable = "Hello" # my_variable is now a string
print("my_variable: ", my_variable, " Type: ", type(my_variable))
```

```python
# Type Checking ans Type Conversion
# You can check the type of a variable using the type() function.
height = 5.6
print(type(height)) 
```

```python
# Dynamic typing allows you to change the type of a variable by assigning a different type of value to it. This flexibility can be useful, but it also requires careful management to avoid unintended consequences in your code.
var = 10;
print(var, type(var))

var = "Now I'm a string"
print(var, type(var))

var = 3.14
print(var, type(var))
```

```python
# simple caluclator
num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))
sum = num1 + num2
print("The sum of ", num1, " and ", num2, " is: ", sum)
difference = num1 - num2
print("The difference of ", num1, " and ", num2, " is: ", difference)
product = num1 * num2
print("The product of ", num1, " and ", num2, " is: ", product)
quotient = num1 / num2  
print("The quotient of ", num1, " and ", num2, " is: ", quotient)
```
