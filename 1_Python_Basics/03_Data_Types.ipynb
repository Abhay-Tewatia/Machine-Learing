```python
# integers
x = 42
print(x);
print(type(x))
```

```python
# floating point data type
y = 3.14
print(y)
print(type(y))
```

```python
# string data type
name  = "Abhay Tewatia"
print(name)
print(type(name))
```

```python
# boolean data type
is_student = True
print(is_student)
print(type(is_student))
```

```python
a = 10
b = 10
type(a==b)
```

```python
# common errors

result = "hello " + str(5)
#  here we are trying to add a string and an integer, which is not allowed in Python. This will raise a TypeError. To fix this, we can either convert the integer to a string or the string to an integer, depending on what we want to achieve. For example:
print("hello " + str(5))  # This will work and output "hello5"
print(result)
```
